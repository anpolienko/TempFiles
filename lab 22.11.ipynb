```python
#вариант 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def orig(x): return np.exp(1/x)
def f(y,x): return -y*np.log(y)/x

def RK(f,y0,x):
    n = len(x); h = (x[-1]-x[0])/n; y = []; y.append(y0)
    for i in range(0,n-1):
        k1 = f(y[i], x[i])
        k2 = f(y[i]+h/2*k1, x[i]+h/2)
        k3 = f(y[i]+h/2*k2, x[i]+h/2)
        k4 = f(y[i]+h*k3, x[i]+h)
        y.append(y[i] + h/6*(k1+2*k2+2*k3+k4))
    return y

y0 = np.exp(1)
x = np.linspace(1,4,20)

y_od = odeint(f,y0,x)

plt.plot(x,orig(x), label = 'solution')
plt.plot(x,RK(f,y0,x), label = 'Runge-Kutta')
plt.plot(x,y_od.flatten(), '-.', label = 'odeint')
plt.legend()
```
