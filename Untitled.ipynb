```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1,2,10); n = len(x)
y = [0]*n; y[0]=2
z = [0]*n; z[0]=1


def dz(x,y,z):
    return 12*y/(x**2) + z*0

def rez(x):
    return x**4 + x**(-3)

for i in range(1,n):
    z[i] = z[i-1] + (x[i]-x[i-1])*dz(x[i-1], y[i-1], z[i-1])
    y[i] = y[i-1] + (x[i]-x[i-1])*z[i-1]

plt.plot(x, rez(x))
plt.plot(x,y)
```
